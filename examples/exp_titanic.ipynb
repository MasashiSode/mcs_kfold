# Test mcs_fold with titanic data

## Data source

- [Kaggle Titanic: Machine Learning from Disaster](https://www.kaggle.com/c/titanic/data?select=train.csv)

## Import liblaries

```python
import pandas as pd
from mcs_kfold import MCSKFold

SEED = 2020

```

## Load example data with pandas

```python
df = pd.read_csv("../tests/test_data/train_titanic.csv")
```

## make instance of `MCSKFold`

```python
num_cv = 5
max_iter = 100
shuffle_mc = True
mcskf = MCSKFold(n_splits=num_cv, max_iter=max_iter, shuffle_mc=shuffle_mc, global_seed=SEED)
```

## set the target and run split

note that we can specify multiple parameters here.

```python
target_cols = ["Survived", "Pclass", "Sex"]
indices = mcskf.split(df=df, target_cols=target_cols)
```

## Now let's see how equally the target parameters distributed!

```python
import matplotlib.pyplot as plt

for fold, (train_index, valid_index) in enumerate(indices):
    df.iloc[train_index][target_cols].hist()

```
